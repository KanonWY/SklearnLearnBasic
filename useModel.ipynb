```python
import joblib
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd


# 测试数据
wine = load_wine()
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)

# 重新划分出新的数据 
# 假设wine是新的data，新给出的数据
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data,wine.target,test_size=0.3)
# Xtrain Ytrain ----->  假设Xtrain表示选出数据元素，Ytrian表示新选出的数据元素的结果

# 从文件中导出模型
clf = joblib.load("save/clf.pkl")

# 使用模型

## 1.获取一个新参数的结果
print("Xtrain[1] = ", Xtrain[1])
print("Ytrain[1] = ", Ytrain[1])
# Xtrain[1]是其中一个参数
res = clf.predict([Xtrain[1]])
print("res = ",res)

## 2.获取一组新参数的结果
print("Ytrain = ",Ytrain)
res2 = clf.predict(Xtrain)
print("res2 = ", res2)

## 3.查看新的数据的得分
score = clf.score(Xtrain, Ytrain)
print("score = ", score)


## 4.绘制预测结果图片
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = [u'SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

# 实际值
acc = Ytrain
# 预测值
pre = clf.predict(Xtrain)

plt.plot(acc,color="r", label="actual value")
plt.plot(pre, color="g", label="predict value")


plt.title("实际值与预测值折线图") 
 
plt.legend()#增加图例
 
plt.show() #显示图片





```
